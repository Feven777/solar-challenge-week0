Outlier Detection

```python

num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
z = np.abs(stats.zscore(df[num_cols], nan_policy="omit"))
z_outliers_mask = (z > 3).any(axis=1)
print("Z-score outlier rows:", int(z_outliers_mask.sum()))
```

```python
# avoid Z-scores  treating zero-at-night as outliers
from scipy.stats import zscore
daytime_mask=(df['GHI']>0) | (df['DHI']>0) | (df['DNI']>0)
day_df=df[daytime_mask]

cols_to_check = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']

z = day_df[cols_to_check].apply(zscore)
outliers = (z.abs() > 3)
print("Outliers per column:\n", outliers.sum())
```

```python
# Replace outliers with median 
for col in cols_to_check:
    median_val = day_df[col].median()
    df.loc[daytime_mask & outliers[col], col] = median_val
```

Time series Analysis

```python

# Load cleaned data
df_cleaned = pd.read_csv("../data/togo_clean.csv")
df_cleaned['Timestamp'] = pd.to_datetime(df_cleaned['Timestamp'])
df_cleaned.set_index('Timestamp', inplace=True)

# Columns to plot individually
cols_to_plot = ['GHI', 'DNI', 'DHI', 'Tamb']

# first 500 rows
subset = df_cleaned.iloc[:500]

# Plot each column separately
for col in cols_to_plot:
    plt.figure(figsize=(18,5))
    plt.bar(subset.index, subset[col], width=0.0005)  
    plt.title(f'Bar Chart of {col} over Time (subset)')
    plt.xlabel('Timestamp')
    plt.ylabel(col)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Group by month and compute average
monthly_avg = df_cleaned.resample('M').mean()

# Plot each variable separately
for col in ['GHI', 'DNI', 'DHI', 'Tamb']:
    plt.figure(figsize=(12,5))
    monthly_avg[col].plot(kind='bar')
    plt.title(f'Monthly Average of {col}')
    plt.ylabel(col)
    plt.xlabel('Month')
    plt.tight_layout()
    plt.show()
```

```python
# Compute average per hour of day
hourly_avg = df_cleaned.groupby(df_cleaned.index.hour).mean()

for col in ['GHI', 'DNI', 'DHI', 'Tamb']:
    plt.figure(figsize=(12,5))
    hourly_avg[col].plot(kind='bar')
    plt.title(f'Average Hourly Pattern of {col} Throughout the Day')
    plt.ylabel(col)
    plt.xlabel('Hour of Day')
    plt.tight_layout()
    plt.show()
```

```python
# Example: simple anomaly detection for GHI
ghi = df_cleaned['GHI']
threshold = ghi.mean() + 3*ghi.std()  # Z-score > 3
anomalies = ghi[ghi > threshold]

plt.figure(figsize=(12,5))
plt.plot(df_cleaned.index, ghi, label='GHI')
plt.scatter(anomalies.index, anomalies, color='red', label='Anomalies')
plt.title('GHI with Anomalies Highlighted')
plt.ylabel('GHI')
plt.xlabel('Timestamp')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Group by Cleaning flag and calculate mean
grouped = df_cleaned.groupby('Cleaning')[['ModA', 'ModB']].mean()
print(grouped)
```

Wind and Distribution analysis

```python
# Bin wind direction into 16 sectors (N, NNE, NE, ...)
bins = np.linspace(0, 360, 17)
labels = [f'{int(b)}' for b in bins[:-1]]
df_cleaned['WD_bin'] = pd.cut(df_cleaned['WD'], bins=bins, labels=labels, include_lowest=True)

# Compute mean wind speed per direction
wind_dir_avg = df_cleaned.groupby('WD_bin')['WS'].mean()

# Radial plot
angles = np.linspace(0, 2*np.pi, len(wind_dir_avg), endpoint=False)
wind_speeds = wind_dir_avg.values

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, polar=True)
bars = ax.bar(angles, wind_speeds, width=0.38, bottom=0.0, alpha=0.7)

# Set the direction labels
ax.set_xticks(angles)
ax.set_xticklabels(labels)
ax.set_title('Wind Rose (Average WS by WD)')
plt.show()
```
